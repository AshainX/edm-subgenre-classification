```python
import os
import shutil
import random
import time

# Define paths
src_folder = '/content/drive/MyDrive/DATASET /NEWDATA/AFROHOUSE'
train_folder = '/content/drive/MyDrive/DATASET /TRAIN'
test_folder = '/content/drive/MyDrive/DATASET /TEST'
#afrohouse is done


# # AMBIENT
# src_folder = '/content/drive/MyDrive/DATASET /NEWDATA/AMBIENT'
# train_folder = '/content/drive/MyDrive/2K500DS/AMBIENT/TRAIN'
# test_folder = '/content/drive/MyDrive/2K500DS/AMBIENT/TEST'

# # BASSHOUSE
# src_folder = '/content/drive/MyDrive/DATASET /NEWDATA/BASSHOUSE'
# train_folder = '/content/drive/MyDrive/2K500DS/BASSHOUSE/TRAIN'
# test_folder = '/content/drive/MyDrive/2K500DS/BASSHOUSE/TEST'

# # BIGROOM
# src_folder = '/content/drive/MyDrive/DATASET /NEWDATA/BIGROOM'
# train_folder = '/content/drive/MyDrive/2K500DS/BIGROOM/TRAIN'
# test_folder = '/content/drive/MyDrive/2K500DS/BIGROOM/TEST'

# # DRUMANDBASS
# src_folder = '/content/drive/MyDrive/DATASET /NEWDATA/DRUMANDBASS'
# train_folder = '/content/drive/MyDrive/2K500DS/DRUMANDBASS/TRAIN'
# test_folder = '/content/drive/MyDrive/2K500DS/DRUMANDBASS/TEST'

# # BASSMUSIC
# src_folder = '/content/drive/MyDrive/DATASET /NEWDATA/BASSMUSIC'
# train_folder = '/content/drive/MyDrive/2K500DS/BASSMUSIC/TRAIN'
# test_folder = '/content/drive/MyDrive/2K500DS/BASSMUSIC/TEST'

# # DUBSTEP
# src_folder = '/content/drive/MyDrive/DATASET /NEWDATA/DUBSTEP'
# train_folder = '/content/drive/MyDrive/2K500DS/DUBSTEP/TRAIN'
# test_folder = '/content/drive/MyDrive/2K500DS/DUBSTEP/TEST'

# # HARDSTYLE
# src_folder = '/content/drive/MyDrive/DATASET /NEWDATA/HARDSTYLE'
# train_folder = '/content/drive/MyDrive/2K500DS/HARDSTYLE/TRAIN'
# test_folder = '/content/drive/MyDrive/2K500DS/HARDSTYLE/TEST'

# # PHONK
# src_folder = '/content/drive/MyDrive/DATASET /NEWDATA/PHONK'
# train_folder = '/content/drive/MyDrive/2K500DS/PHONK/TRAIN'
# test_folder = '/content/drive/MyDrive/2K500DS/PHONK/TEST'

# # PSYTRANCE
# src_folder = '/content/drive/MyDrive/DATASET /NEWDATA/PSYTRANCE'
# train_folder = '/content/drive/MyDrive/2K500DS/PSYTRANCE/TRAIN'
# test_folder = '/content/drive/MyDrive/2K500DS/PSYTRANCE/TEST'


# Create TRAIN and TEST folders if they don't exist
os.makedirs(train_folder, exist_ok=True)
os.makedirs(test_folder, exist_ok=True)

# Get all MP3 files in the source folder
mp3_files = [f for f in os.listdir(src_folder) if f.endswith('.mp3')]

# Shuffle the list of MP3 files randomly
random.shuffle(mp3_files)

# Split into TRAIN and TEST sets
train_files = mp3_files[:2000]
test_files = mp3_files[2000:]

# Copy files to respective folders with error handling and logging
for i, file in enumerate(train_files, start=1):
    try:
        shutil.copy(os.path.join(src_folder, file), train_folder)
        if i % 100 == 0:
            print(f"{i} files copied to TRAIN folder.")
        time.sleep(0.01)  # Small delay to prevent Google Drive throttling
    except Exception as e:
        print(f"Error copying {file} to TRAIN: {e}")

for i, file in enumerate(test_files, start=1):
    try:
        shutil.copy(os.path.join(src_folder, file), test_folder)
        if i % 100 == 0:
            print(f"{i} files copied to TEST folder.")
        time.sleep(0.01)  # Small delay to prevent Google Drive throttling
    except Exception as e:
        print(f"Error copying {file} to TEST: {e}")

print(f"Copied {len(train_files)} files to TRAIN folder.")
print(f"Copied {len(test_files)} files to TEST folder.")
```
